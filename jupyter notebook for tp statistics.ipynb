```python
import statistics as st
import numpy as np
t=[1,2,4,1,6,3,2,1,2,0]
print("tableau",t)
moy=sum(t)/len(t)
print("moyenne : ",moy)
var = sum((x-moy)**2 for x in t) / len(t)
print("variance",var)
t.sort()
print("tableau trié",t)
if(len(t)%2==0):
    me=t[(len(t)//2)-1]
else:
    me=(t[(len(t)//2)-1]+t[(len(t)//2)])/2
print("mediane par nos algo",me)
```

```python
print("_______avec numpy________")
t=[1,2,4,1,6,3,2,1,2,0]
print("variance: ", np.var(t))
print("mediane ",np.median(t))
print("ecartype ",np.std(t))
print("quantile 25%",np.quantile(t,0.25))
print("quantile 75%",np.quantile(t,0.75))
print("min : ",min(t))
print("max : ",max(t))
a,b=np.unique(t,return_counts=True)
print("unique values    : ",a)
print("unique frequence : ",b)
```

```python
print("_______avec statistics________")
print("variance", st.variance(t))
print("quantile",st.quantiles(t))
t=[1,2,4,1,6,3,2,1,2,0]
print("mode",st.multimode(t))
```

```python
def density(x):
    return (1/(np.sqrt(2*np.pi))*np.exp(-(0.5)*x**2))
```

```python
y=np.random.normal(0,1,1000)
plt.hist(y,density=1)
x=np.linspace(-3,3,200)
plt.plot(x,density(x))
```
