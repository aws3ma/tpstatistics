```python
xiArg,niArg=np.unique(data.Number_of_days,return_counts=True)
print(*xiArg)
print(*niArg)
```

```python
sommeCumuléArg=np.cumsum(niArg)
print(*sommeCumuléArg)
```

```python
data.Number_of_days.std()
```

```python
a=niArg/sommeCumuléArg[len(sommeCumuléArg)-1]
freqcum=np.cumsum(a)
mondf=pd.DataFrame([xiArg,niArg,sommeCumuléArg,a,freqcum],index=["Number_of_days","effectif","effectif cumulé","frequence","frequence cumulé"])
Tab=mondf.T
print("Tableau de employe par nombre du jour =")
Tab
```
